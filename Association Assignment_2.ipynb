### Using the Dataset 2. my_movies.csv

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
movies=pd.read_csv("F:/ExcelR workings/Assignments/Association Rules/my_movies.csv")
```

```python
movies.head()
```

```python
data=movies.iloc[:,5:15]
```

```python
data.head()
```

# Apriori Algorithm

```python
frequent_itemsets=apriori(data, min_support=0.1,max_len=5, use_colnames=True)
frequent_itemsets
```

```python
frequent_itemsets
```

```python
# Setup the rules
rules=association_rules(frequent_itemsets, metric="lift", min_threshold=0.7)
rules
```

```python
#Lets Sort the Rules By lift
rules.sort_values('lift',ascending = False)[0:20]
```

```python
rules[rules.lift>1]
```

### Visualization using plots

```python
import matplotlib.pyplot as plt
plt.scatter(rules['support'], rules['confidence'], alpha=0.5)
plt.xlabel('support')
plt.ylabel('confidence')
plt.title('Support vs Confidence')
plt.show()
```

```python
plt.scatter(rules['support'], rules['lift'], alpha=0.5)
plt.xlabel('support')
plt.ylabel('lift')
plt.title('Support vs Lift')
plt.show()
```

```python
import numpy as np
fit=np.polyfit(rules['lift'],rules['confidence'],1)
fit_fn=np.poly1d(fit)
plt.plot(rules['lift'],rules['confidence'], 'yo', rules['lift'], fit_fn(rules['lift']))
```
