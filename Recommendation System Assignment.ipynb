# Build a recommender system by using cosine simillarties score.



```python
import pandas as pd
import numpy as np 
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
book=pd.read_csv("F:/ExcelR workings/Assignments/Recommendation system/book.csv", engine='python')
```

```python
book[0:10]
```

```python
data=book.copy()
```

```python
Book=data.drop(['Unnamed: 0'],axis=1)
```

```python
Book.rename({'User.ID':'ID','Book.Title':'BT','Book.Rating':'R'}, axis=1, inplace=True)
Book.head()
```

```python
# No. of unique users in the dataset
len(Book.ID.unique())
```

```python
len(Book.BT.unique())
```

```python
avg_range=pd.DataFrame(Book.groupby('BT')['R'].mean())
avg_range.head()
```

```python
avg_range['TR']=pd.DataFrame(Book.groupby('BT')['R'].count())
avg_range.head()
```

```python
avg_range.shape
```

```python
B_user=Book.pivot_table(index='ID', columns='BT', values='R')
B_user.head()
```

```python
B_user.index= Book.ID.unique()
```

```python
B_user
```

```python
# Impute those Nans with 0 values
B_user.fillna(0, inplace=True)
```

```python
B_user
```

### Calculating Cosine Similarity between users.

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim= 1-pairwise_distances(B_user.values, metric= 'cosine')
```

```python
user_sim
```

```python
# Store the results in a dataframe
user_sim_df=pd.DataFrame(user_sim)
```

```python
# Set the index and column names to user ids
user_sim_df.index=Book.ID.unique()
user_sim_df.columns=Book.ID.unique()
```

```python
user_sim_df
```

```python
user_sim_df.iloc[0:10,0:10]
```

```python
# Most similar users
user_sim_df.idxmax(axis=1)[0:10]
```
